**Goal today:** Build a real-world preprocessing pipeline on SMS spam data so it becomes ML-ready.

## Step 1: Import libraries
We start by importing Python libraries needed for data handling and NLP preprocessing.

```python
import os
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

## Step 2: Download NLTK resources locally
We save NLTK data inside the project folder so the notebook works reliably across systems.

```python
LOCAL_NLTK_DIR = os.path.join(os.getcwd(), '.nltk_data')
os.makedirs(LOCAL_NLTK_DIR, exist_ok=True)
if LOCAL_NLTK_DIR not in nltk.data.path:
    nltk.data.path.insert(0, LOCAL_NLTK_DIR)

nltk.download('punkt', quiet=True, download_dir=LOCAL_NLTK_DIR)
nltk.download('punkt_tab', quiet=True, download_dir=LOCAL_NLTK_DIR)
nltk.download('stopwords', quiet=True, download_dir=LOCAL_NLTK_DIR)

print('NLTK resources ready at:', LOCAL_NLTK_DIR)
```

## Step 3: Load dataset
The SMS Spam dataset has labels and message text. We keep only required columns.

```python
df = pd.read_csv('spam.csv', encoding='latin-1')

df = df[['v1', 'v2']].copy()
df.columns = ['label', 'text']

df.head()
```

## Step 4: Build text cleaning function
We convert to lowercase, tokenize, remove punctuation/noise, and remove stopwords.

```python
def clean_text(text: str):
    text = str(text).lower()
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word.isalnum()]
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return tokens
```

## Step 5: Apply preprocessing to entire dataset
This creates a new column with cleaned tokens for each message.

```python
df['cleaned_text'] = df['text'].apply(clean_text)

df[['text', 'cleaned_text']].head()
```

## Step 6: Inspect data and class balance
In spam detection, class balance matters a lot for modeling quality.

```python
print('Total messages:', len(df))
print('\nClass distribution:')
print(df['label'].value_counts())
```

## Step 7: Convert labels to numeric
Machine learning models work with numeric labels.

- `ham -> 0`
- `spam -> 1`

```python
df['label_num'] = df['label'].map({'ham': 0, 'spam': 1})

df[['label', 'label_num']].head()
```
