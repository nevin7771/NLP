# Week 1 - Day 4: Streamlit Web App

Day 4 turns your NLP pipeline into something **shareable**: a small web UI where anyone can paste text and see **HAM vs SPAM** plus **probabilities**.

## What’s in the repo
- `app.py` — Streamlit app (TF-IDF + balanced Logistic Regression, trained on full `spam.csv` for the demo)
- `nlp_utils.py` — same preprocessing as Days 1–3 (`clean_text`, NLTK setup)

## Run the app (from this project folder)
```bash
uv sync
uv run streamlit run app.py
```

Then open the URL Streamlit prints (usually `http://localhost:8501`).

## What you’ll demo
- Paste SMS-style text → click **Predict**
- See label (**HAM** / **SPAM**) and probability for each class
- Optional: screenshot for LinkedIn / portfolio

## Note
The first load trains the model once and caches it (Streamlit `@st.cache_resource`). In production you’d load **pickled** `vectorizer + model` instead of training on startup.
